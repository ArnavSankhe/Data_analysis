```python
import datetime
def from_dob_to_age(born):
    today = datetime.date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
master['birth_date'] = master['birth_date'].apply(lambda x: from_dob_to_age(x))
```
